#### Observation
- For handling categorical columns like "Mocodes", "Vict Sex", and "Vict Descent" we use the high frequency values in the column to fill in the null values.

- Similarly for handling numerical columns like "Premis Cd" and "Crm Cd 1" we use the high frequency values in the column to fill in the null values.

- For the rest of the columns there's more than 50% missing values, which are columns "Weapon Used Cd", "Weapon Desc", "Crm Cd 2   ", "Crm Cd 3", "Crm Cd 4" and "Cross Street",  we can drop the columns since they are redundant data and replacing the missing values in those columns with any data handling would cause some miscellaneous results, which is why we are dropping them off.

```python
# Handle missing data
# Handling categorical columns
df['Mocodes'].fillna(df['Mocodes'].mode()[0], inplace=True)
df['Vict Sex'].fillna(df['Vict Sex'].mode()[0], inplace=True)
df['Vict Descent'].fillna(df['Vict Descent'].mode()[0], inplace=True)

# Handling numerical columns
df['Premis Cd'].fillna(df['Premis Cd'].mode()[0], inplace=True)
df['Crm Cd 1'].fillna(df['Crm Cd 1'].mode()[0], inplace=True)

# Handling text columns
df['Premis Desc'].fillna(df['Premis Desc'].mode()[0], inplace=True)
```

```python
# Drop specified columns
columns_to_drop = ['Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']
df.drop(columns_to_drop, axis=1, inplace=True)

# Verify if columns have been dropped
print("Columns after dropping specified columns: \n", df.columns)
```

```python
from datetime import datetime
from datetime import timezone
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])

# Custom function to pad single-digit hours and minutes
def pad_time(time_str):
    time_str = str(time_str).zfill(4)
    return f"{time_str[:2]}:{time_str[2:]}"

# Apply the custom function and convert to datetime
df['TIME OCC'] = df['TIME OCC'].apply(pad_time)
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M')
```

```python
#4.7 Impact of major events
df_events = pd.read_csv("Major events file.csv")
df['YEAR'] = df['DATE OCC'].dt.year
merged_data = pd.merge(df, df_events, on='YEAR', how='inner')
def analyze_crime_rate_change(event, window_years=2):
    event_year = event['YEAR']
    before_event = merged_data[(merged_data['YEAR'] >= event_year - window_years) & (merged_data['YEAR'] < event_year)]
    after_event = merged_data[(merged_data['YEAR'] > event_year) & (merged_data['YEAR'] <= event_year + window_years)]
    crime_rate_before = before_event['crime_rate'].mean()
    crime_rate_after = after_event['crime_rate'].mean()

    return {
        'Event': event['Event'],
        'Event Year': event_year,
        'Crime Rate Before Event': crime_rate_before,
        'Crime Rate After Event': crime_rate_after
    }
analysis_results = []
for index, event in df_events.iterrows():
    result = analyze_crime_rate_change(event)
    analysis_results.append(result)
results_df = pd.DataFrame(analysis_results)
```

```python
from pmdarima.arima import auto_arima
df3 = pd.DataFrame(df.groupby(['Month','Year']).size()).reset_index()
df3['Month_Year'] = df3.apply(lambda row: str(row['Year']) + ' - ' + str(row['Month']), axis=1)
df3.columns.values[2] = 'Count'
df3.sort_values(by = ['Year','Month'], inplace=True)

data = df3.loc[:,['Month_Year','Count']]
data.set_index('Month_Year', inplace=True)

# Specify the maximum values for p, d, q, P, D, Q, and the seasonal period S
# Setting them to None means the function will search over a range of values.
model = auto_arima(data, seasonal=True, stepwise=True, trace=True,
                  suppress_warnings=True, max_order=None)
```

```python
print(model.summary())
print(model.order)
print(model.seasonal_order)
```

```python
predictions = model.predict_in_sample()
```

```python
plt.figure(figsize = (15,6))
plt.plot(data.index, data['Count'], label='Actual Data', marker='o')
plt.plot(data.index, predictions, label='Predicted Data', marker='x', linestyle='--', color='orange')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title(' Distribution of actual crime data vs predicted data')
plt.legend(loc = 'center right')
plt.xticks(rotation = 90)
plt.show()
```
